```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

pipe.enable_attention_slicing()
```

# Generate synthetic images for glacier class

```python
from PIL import Image
import os

os.makedirs("/content/glacier_images", exist_ok=True)

prompts = [
    "glacier landscape, ice and snow, rocky terrain, natural photography, overcast sky",
    "glacier valley, blue ice, snow covered rocks, mountain background, realistic photo",
    "glacial ice formation, frozen landscape, grey sky, nature photography",
    "glacier surface, cracked blue ice, snow, cold weather, daylight, photorealistic",
    "arctic glacier, ice field with rocks, misty mountains, natural light",
    "glacier moraine, ice and gravel, cloudy sky, remote wilderness photo",
    "glacial lake, ice chunks, snow mountains background, cold blue tones",
    "glacier edge, melting ice, rocky ground, overcast lighting, nature scene",
]

negative_prompt = (
    "people, buildings, roads, trees, forest, warm colors, "
    "tropical, desert, cartoon, painting, aerial view, satellite, "
    "low quality, blurry, watermark"
)

for i in range(220):
    image = pipe(
        prompts[i % len(prompts)],
        negative_prompt=negative_prompt,
        width=512,
        height=512,
        num_inference_steps=35,
        guidance_scale=8.0,
        generator=torch.manual_seed(i)
    ).images[0]

    image = image.resize((150, 150), Image.LANCZOS)
    image.save(f"/content/glacier_images/glacier_{i:03d}.png")
    print(f"{i+1}/220 Done")

print("Completed!")
```

```python
import numpy as np
from PIL import Image
import os

imgs = []
for f in os.listdir("/content/glacier_images"):
    img = np.array(Image.open(f"/content/glacier_images/{f}")) / 255.0
    imgs.append(img)

imgs = np.stack(imgs)
print(f"Mean — R: {imgs[:,:,:,0].mean():.4f}, G: {imgs[:,:,:,1].mean():.4f}, B: {imgs[:,:,:,2].mean():.4f}")
print(f"Std  — R: {imgs[:,:,:,0].std():.4f}, G: {imgs[:,:,:,1].std():.4f}, B: {imgs[:,:,:,2].std():.4f}")
```

# Generate synthetic images for forest

```python
prompts = [
    "dense forest, tall trees, green foliage, natural photography, sunlight through leaves",
    "forest landscape, coniferous trees, pine forest, overcast sky, nature photo",
    "woodland scene, deciduous forest, green canopy, realistic photography",
    "forest path, trees and undergrowth, dappled sunlight, nature photography",
    "dense jungle forest, tropical trees, lush green, natural light, photorealistic",
    "forest interior, tree trunks, moss, ferns, soft lighting, nature scene",
    "autumn forest, colorful leaves, trees, natural photography, warm tones",
    "misty forest, fog between trees, green trees, atmospheric, nature photo",
    "forest canopy, looking up at trees, green leaves, sunlight, realistic",
    "boreal forest, spruce trees, wilderness, overcast, remote nature photography",
]

negative_prompt = (
    "people, buildings, roads, cars, cartoon, painting, "
    "aerial view, satellite, low quality, blurry, watermark, "
    "glacier, snow, ice, desert, ocean"
)
```

```python
os.makedirs("/content/forest_images", exist_ok=True)

for i in range(220):
    image = pipe(
        prompts[i % len(prompts)],
        negative_prompt=negative_prompt,
        width=512,
        height=512,
        num_inference_steps=35,
        guidance_scale=8.0,
        generator=torch.manual_seed(i)
    ).images[0]

    image = image.resize((150, 150), Image.LANCZOS)
    image.save(f"/content/forest_images/forest_{i:03d}.png")

    if i % 20 == 0:
        print(f"{i}/220 Done")

print("Completed!")
```
