```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image_dir = "synthetic_data/glacier_images"
image_files = os.listdir(image_dir)
```

```python
selected = random.sample(image_files, 10)

fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for ax, fname in zip(axes.flat, selected):
    img = Image.open(os.path.join(image_dir, fname))
    ax.imshow(img)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
